### Plot horizontal bars of "tamponi_test_antigenico_rapido" and "tamponi_test_molecolare" columns for each region in different subplots

`hint: df[].plot.barh, figsize=()`

- Figure size of (6,10)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(len(df), 1, figsize=(6, 10*2))

for i, regione in enumerate(df.index):
    # Estrai i dati della regione corrente
    dati_regione = df.loc[regione, ['tamponi_test_antigenico_rapido', 'tamponi_test_molecolare']]

    # Creo un DataFrame temporaneo per il plot con nomi più leggibili
    df_plot = pd.DataFrame({
        'Antigenico Rapido': [dati_regione['tamponi_test_antigenico_rapido']],
        'Molecolare': [dati_regione['tamponi_test_molecolare']]
    })
    df_plot.plot.barh(ax=axs[i])
    axs[i].set_title(f'Regione: {regione}')
    axs[i].set_xlabel('Numero tamponi')
    axs[i].grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

### Plot horizontal bars of "tamponi_test_antigenico_rapido" and "tamponi_test_molecolare" columns for each region sorting by "tamponi_test_antigenico_rapido" column

### Plot histograms of "totale_ospedalizzati", "terapia_intensiva", "ricoverati_con_sintomi", "nuovi_positivi" in different subplots

`hint: df[].plot.hist, bins=, alpha=`

- Figure size of (10, 6)
- 10 Bins
- Alpha of 0.75

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

df["totale_ospedalizzati"].plot.hist(ax=axes[0,0], bins=10, alpha=0.75)
axes[0,0].set_title("totale_ospedalizzati")

df["terapia_intensiva"].plot.hist(ax=axes[0,1], bins=10, alpha=0.75)
axes[0,1].set_title("terapia_intensiva")

df["ricoverati_con_sintomi"].plot.hist(ax=axes[1,0], bins=10, alpha=0.75)
axes[1,0].set_title("ricoverati_con_sintomi")

df["nuovi_positivi"].plot.hist(ax=axes[1,1], bins=10, alpha=0.75)
axes[1,1].set_title("nuovi_positivi")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
regions_colors = { "Abruzzo": "orange", "Basilicata": "orange", "Calabria": "orange", "Campania": "red",
                  "Emilia-Romagna": "red", "Friuli Venezia Giulia": "red", "Lazio": "red", "Liguria": "orange",
                   "Lombardia": "red", "Marche": "red", "Molise": "red", "P.A. Bolzano": "orange",
                   "P.A. Trento": "red", "Piemonte": "red", "Puglia": "red", "Sardegna": "white", "Sicilia": "red",
                   "Toscana": "orange", "Umbria": "orange", "Valle d'Aosta": "orange", "Veneto": "red",
}
color_series = pd.Series(regions_colors)
df['color'] = color_series
df.groupby('color')['nuovi_positivi'].mean().plot.bar(figsize=(8, 6), rot=45)
plt.show()
```

### Group the regions by color and plot bars of mean *and error* of "nuovi_positivi" column

`hint: df.groupby().std(), yerr=, rot=,`

```python
df.groupby('color')['nuovi_positivi'].agg(['mean', 'std']).plot.bar(yerr='std', figsize=(8, 6), rot=45, capsize=4)
plt.show()
```
